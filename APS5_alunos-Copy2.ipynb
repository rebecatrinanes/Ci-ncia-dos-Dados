Faça o fit de uma distribuição exponencial sobre os dados da Magnitude, achando os valores de **loc** e **scale**. Interprete loc e scale no caso da exponencial.
Documentação: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html

```python
#loc é média
#scale é desvio padrão

mu=df["Magnitude"].mean()
sigma=df["Magnitude"].std()
print(mu)
print(sigma)
```

Refaça o Histograma plotando a fdp (função densidade de probabilidade) da exponencial com os parâmetros achados no fit em cima. Cuidado com o domínio utilizado. Interprete.

```python
from scipy import stats

x=sorted(stats.norm.rvs(size=10000, loc=mu, scale=sigma))
faixa=np.arange(5.5,9,0.1)
plot = df.Magnitude.plot.hist(bins=faixa,title='Magnitude', normed=1, alpha=0.8,color="r",sharex=True, sharey=True,histtype='stepfilled')
plt.ylabel('Densidade')
plt.xlabel('Magnitude')
plt.grid()
plt.plot(x, expon.pdf(x, loc=mu, scale=sigma), lw=3, alpha=0.9, color="c")
plt.show()
```

O fit faz com que a média seja o pico da exponencial (5.88) e decaia de acordo com seu desvio padrão de 0.42 ## MUDAR

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#Se usar o arquivo descompactado
#pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
dados2015 = pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
dados2014 = pd.read_csv('DOM2014csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
dados2015
```

```python
#delimitando a tabela com os dados que eu quero
dados2015 = dados2015.loc[(dados2015.UF==29) & (dados2015.V0202==2) & (dados2015.V4105==1),["UF","V0202","V0207","V4105","V0226","V02273","V02274","V4614"]]

#transformando int em categoria
dados2015.UF = dados2015.UF.astype("category")
dados2015.V0202 = dados2015.V0202.astype("category")
dados2015.V0207 = dados2015.V0202.astype("category")
dados2015.V4105 = dados2015.V4105.astype("category")
dados2015.V0226 = dados2015.V0226.astype("category")
dados2015.V02273 = dados2015.V02273.astype("category")
dados2015.V02274 = dados2015.V02274.astype("category")


#renomeando as linhas
dados2015.UF.cat.categories = ("Bahia")
dados2015.V0202.cat.categories = ("Casa")
dados2015.V0207.cat.categories = ("Próprio")
dados2015.V4105.cat.categories = ("Urbana")
dados2015.V0226.cat.categories = ("Sim","Não")
dados2015.V02273.cat.categories = ("Sim","Não","Não Sabe")
dados2015.V02274.cat.categories = ("Sim","Não","Não Sabe")

#renomemando as colunas
dados2015 = dados2015.rename(columns={"UF":"REGIAO","V4614": "RENDA","V0202":"DOMICILIO","V0207":"CONDOCUP","V4105":"AREA","V0226":"TVCOR","V02273":"TVABERTA","V02274":"TVANTENA"})

dados2015 = dados2015[pd.notnull(dados2015["TVABERTA"])]
dados2015 = dados2015[pd.notnull(dados2015["TVANTENA"])]

dados2015                              
               
```

```python
'''import matplotlib.pyplot as plt
import matplotlib

tv_cor = (dados2015.TVCOR.value_counts(sort=False, normalize=True)*100).round(decimals=1)

tv_aberta = (dados2015.TVABERTA.value_counts(sort=False, normalize=True)*100).round(decimals=1)

tv_antena = (dados2015.TVANTENA.value_counts(sort=False, normalize=True)*100).round(decimals=1)

                         
fig = plt.figure(figsize=(15, 5))
plt.subplot(131)
plot = tv_cor.plot(kind='pie',title='TV c/ cor',autopct='%.1f')

plt.subplot(132)
plot = tv_aberta.plot(kind='pie',title='TV aberta',autopct='%.1f')

plt.subplot(133)
plot = tv_antena.plot(kind='pie',title='TV c/ antena',autopct='%.1f')

plt.tight_layout()
plt.show()'''
```

```python
dados20151 = dados2015[(dados2015.RENDA >= 100) & (dados2015.RENDA < 1000)]

tv_cor = (dados20151.TVCOR.value_counts(sort=False, normalize=True)*100).round(decimals=1)

tv_aberta = (dados20151.TVABERTA.value_counts(sort=False, normalize=True)*100).round(decimals=1)

tv_antena = (dados20151.TVANTENA.value_counts(sort=False, normalize=True)*100).round(decimals=1)

                         
fig = plt.figure(figsize=(15, 5))
plt.subplot(131)
plot = tv_cor.plot(kind='pie',title='TV c/ cor',autopct='%.1f')

plt.subplot(132)
plot = tv_aberta.plot(kind='pie',title='TV aberta',autopct='%.1f')

plt.subplot(133)
plot = tv_antena.plot(kind='pie',title='TV c/ antena',autopct='%.1f')

plt.tight_layout()
plt.show()
```
